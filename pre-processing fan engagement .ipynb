```python
#Basic libraries
import pandas as pd 
import numpy as np 
#Ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
raw_reviews = pd.read_csv('data.csv', parse_dates=['DataCollectedTimeStamp'])
```

```python
print (raw_reviews.info())
```

```python
raw_reviews.head()
```

```python
raw_reviews['DataCollectedTimeStamp'] = pd.to_datetime(raw_reviews.DataCollectedTimeStamp,errors='coerce')
```

```python
print (raw_reviews.info())
```

```python
raw_reviews.set_index('DataCollectedTimeStamp')
```

```python
raw_reviews.groupby(raw_reviews['DataCollectedTimeStamp']).describe()
```

```python
df = raw_reviews.groupby(raw_reviews['DataCollectedTimeStamp']).mean()
```

```python
df.to_csv('out.csv')
```
